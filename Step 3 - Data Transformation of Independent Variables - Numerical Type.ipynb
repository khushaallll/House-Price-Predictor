# Conversion of a categorical variable to numerical variable
No of Times Visited

```python
data['No of Times Visited'].unique()
```

```python
'''
No of Times Visited - is a categorical variable but if we convert once twice
etc to 1 , 2 ... we can convert it to numerical type and thus establish a 
possible relationship between "No of Times Visited" and "Sale Price"
We can use map function to do so.

'''

mappings = { 'None':'0', 'Thrice' : '3', 'Four':'4', 'Twice':'2', 'Once':'1'}
data['No of Times Visited'] = data['No of Times Visited'].map(mappings)
```

```python
data['No of Times Visited'].unique()
```

# Deriving two variables from Existing variable
We derive from "Renovated Year"

```python
#Now, we can combine two variables to make a useful variable
# Or we can derive a variable from existing variable as well.
# we derive two variables from existing variable- "Renovated Year"
# 1. Ever renovated
# 2. Years Since Renovation
```

```python
# if Renovated Year is 0 then put No else Yes
data['Ever Renovated'] = np.where(data['Renovated Year']== 0, 'No', 'Yes')
```

```python
#Years Since Renovation = Year of Sale - Year of Renovation
data['Purchase Year'] =pd.DatetimeIndex(data['Date House was Sold']).year
data['Years Since Renovation'] = np.where(data['Ever Renovated']=='Yes',
                                        abs(data['Purchase Year']-data['Renovated Year']),0)
```

```python
#We can drop certain columns
data.drop(columns=['Date House was Sold','Renovated Year','Purchase Year', 'Unnamed: 0','Unnamed: 0.1'],inplace=True)
```
