# Treating Missing Values in Independent Variable by Imputing method

```python
#zipcode is not included because it is a categorical variable.
#replacing the missing value by calculating median of all zipcodes does not make sense.
num_col = ['No of Bathrooms','Flat Area (in Sqft)','Lot Area (in Sqft)',
           'Area of the House from Basement (in Sqft)',
          'Latitude','Longitude','Living Area after Renovation (in Sqft)']
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'median')
data[num_col] = imputer.fit_transform(data[num_col])
```

```python
data.info()
```

```python
data['Zipcode'].shape
```

```python
col = data['Zipcode'].values.reshape(-1,1)
col.shape
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
data['Zipcode'] = imputer.fit_transform(col)
```
